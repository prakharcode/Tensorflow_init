## Simple Neural Net Classifier from Scratch

```python
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.metrics import confusion_matrix 
import numpy as np
from __future__ import print_function 
```

## Load MNIST data

```python
from tensorflow.examples.tutorials.mnist import input_data
data = input_data.read_data_sets("data/MNIST", one_hot = True)
```

## Data Dimensions and Plotting

```python
print("The length of train set: {0}".format(data.train.labels.size))
print("The length of validation set: {0}".format(data.validation.labels.size))
print("The length of test set: {0}".format(data.test.labels.size))
```

```python
# total no of pixel in horizontal and vertical direction each
pixel = 28
img_shape = (28, 28)

# total no of pixels

tot_pixels = 28 * 28
tot_classes = 10
print("Total number of pixels: {0}".format(tot_pixels))
print("Total Class: {0}".format(tot_classes))
print("Shape of the image:{0}".format(img_shape))
```

```python
def plot_image(plot_img, plot_cls):  # function to plot multiple images
    tot_plot_img = len(plot_img)
    plot_rows = 2
    plot_cols = tot_plot_img/ plot_rows if tot_plot_img % 2 == 0 else (tot_plot_img / plot_rows)+1
    plt.figure(1)
    for i in range(tot_plot_img):
        plt.subplot(plot_rows, plot_cols, i+1)
        plt.axis('off')
        plt.imshow(plot_img[i].reshape(img_shape), cmap='gray')
        plt.title("True: {0}".format(plot_cls[i].tolist().index(1)))
    plt.subplots_adjust(right = 1, wspace=0.3)
```

```python
plot_image(data.train.images[:10], data.train.labels[:10])
```

# Tensorflow Graph

Tensorflow graph can be thought of as a wiring of the complete neural network done before a current(data) is made to flow through the wiring. It's easy to draw a graph and then see it through breadboard or tensorboard.

## Initializing Placeholders

```python
## Placeholder Variables
X = tf.placeholder(tf.float32, shape =(None, tot_pixels), name = "X")
Y = tf.placeholder(tf.float32, shape =(None, tot_classes), name="Y")
```

## Initializing Weights and Biases

```python
## Variables initialization
initializer = tf.contrib.layers.xavier_initializer()
W1 = tf.Variable(initializer([tot_pixels, 100]), name = "Weight_Layer_1")
W2 = tf.Variable(initializer([100, 50]), name = "Weight_Layer_2")
W3 = tf.Variable(initializer([50, tot_classes]), name = "Weight_Layer_3")

b1 = tf.Variable(tf.zeros([100]), name = "bias_1")
b2 = tf.Variable(tf.zeros([50]), name = "bias_2")
b3 = tf.Variable(tf.zeros([tot_classes]), name = "bias_3")            
```

## Affine Layer Calculation

The whole graph is calculated through these Affine transformations

```python
## Affine Layers
layer_1 = tf.nn.sigmoid(tf.matmul(X, W1) + b1, name="Affine_1")

layer_2 = tf.nn.sigmoid(tf.matmul(layer_1, W2) + b2, name="Affine_2")

logits = tf.add(tf.matmul(layer_2, W3) , b3, name="Affine__3")
```

## Predictions and labels

```python
## Prediction 
y_pred = tf.nn.softmax(logits, name='Last_activation')
y_pred_class = tf.argmax(y_pred, axis=1)
y_true_class = tf.argmax(Y, axis=1)
```

## Cost function and Optimizer function

```python
## cost function and optimizer
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits = logits, labels = Y), name='Cost_function')
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.4, name='SGD_Optimizer').minimize(cost)

# summary for graph
tf.summary.scalar("loss", cost)
```

## Accuracy evaluation

```python
correct_prediction = tf.equal(y_pred_class, y_true_class)
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
# summary for graph

tf.summary.scalar("accuracy", accuracy)

merged_summary_op = tf.summary.merge_all()
```

## Running the session 

After completng all the required variables and function you have to initialize all the variable which is equivalent to creating a graph.

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
writer = tf.summary.FileWriter("../deeplearning/ann-check/", sess.graph)
```

## Optimizing function

```python
def optimize(num_iter, batch_size = 64):
    for i in range(num_iter):
        x_batch, y_true_batch = data.train.next_batch(batch_size)
        if(i % (num_iter/100)==0):
            print('|', end='')
        feed_dict_train = {X : x_batch, Y:y_true_batch }
        _, c, summary = sess.run([optimizer, cost, merged_summary_op], feed_dict = feed_dict_train)
        writer.add_summary(summary)
    print('', end='\n')
```

## Helper function to calculate the accuracy

```python
def print_acc():
    feed_dict_test = { X: data.test.images, Y:data.test.labels}
    acc = sess.run(accuracy, feed_dict=feed_dict_test)
    print('{0:.2%}'.format(acc))
```

#### Accuracy before any optimization

```python
print_acc()
```

#### Running Optimization

```python
optimize(100000)
```

#### Accuracy after running optimization

```python
print_acc()
```

## Printing Confusion Matrix

Another way to visualize your classifier's output is by using a plotting technique called confusion matrix plot

```python
def conf_mat():
    feed_dict_test = { X: data.test.images, Y:data.test.labels}
    y_true, y_pred = np.argmax(data.test.labels, axis = 1),sess.run(y_pred_class, feed_dict=feed_dict_test)
    cm = confusion_matrix(y_true, y_pred)
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Reds)
    plt.tight_layout()
    plt.colorbar()
    tick_marks = np.arange(tot_classes)
    plt.xticks(tick_marks, range(tot_classes))
    plt.yticks(tick_marks, range(tot_classes))
    plt.xlabel('Predicted')
    plt.ylabel('True')
    
    plt.show()
```

```python
conf_mat()
```
